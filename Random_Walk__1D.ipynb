```
import numpy as np
import random as random
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

## **One dimensional (1D) random walk :** 

```
P = [] #empty list to collect position of walk
T = [] #empty list to collect time of walk
x = 0  #starting point origin
d = 5  #step length
n = 1000 #no.of steps
for r in range(n): #using for loop
    x = x + d*random.choice([1, -1])
    T.append(r)
    P.append(x)
#print(P)
#print(T)
```

```
plt.figure(figsize = [10,4])
plt.plot(T,P, marker = 'o', color = 'r', lw = 3)
plt.xlabel('Time', fontsize = 25)
plt.ylabel('Position', fontsize = 25)
plt.xticks(fontsize = 20, color = 'g')
plt.yticks(fontsize = 20, color = 'g')
plt.show()
```

## One dimensional random walk using function.

```
def walk1D(x,n,d):
    P = [] 
    T = []
    for r in range(n):
        x = x + d*random.choice([-1, 1])
        T.append(r)
        P.append(x)
    return P,T
P1,T1 = walk1D(x = 0, n = 1000, d = 5) #for first particle
P2,T2 = walk1D(x = 0, n = 1000, d = 5) #for second particle
#print(P1, T1)
#print(P2,T2)
```

```
plt.figure(figsize = [10,3])
plt.plot(T1, P1, '-', 'r', lw = 2)
plt.plot(T2, P2, '-', 'b', lw = 2)
plt.xlabel('Time', fontsize = 25)
plt.ylabel('Position', fontsize = 25)
plt.xticks(fontsize = 20, color = 'g')
plt.yticks(fontsize = 20, color = 'g')
plt.show()
```

## One dimensional random walk using class

```
class Walk1D(object):
    def __init__ (self, n, d, x0):
        self.n = n
        self.P = [] #empty set to collect position
        self.d = d
        self.x0 = x0
    def walk1D(self):
        x = self.x0
        m = 0
        while m<n:
            x = x + self.d*random.choice([1,-1])
            self.P.append(x)
            m = m + 1
        return self.P
```

```
P1 = Walk1D(n = 1000, d = 5, x0 = 0).walk1D()
P2 = Walk1D(n = 1000, d = 5, x0 = 0).walk1D()
P3 = Walk1D(n = 1000, d = 5, x0 = 0).walk1D()
P4 = Walk1D(n = 1000, d = 5, x0 = 0).walk1D()
T = [r for r in range(len(P1))]
#print(len(P1))
```

```
plt.figure(figsize = [11,5])
plt.plot(T, P1, '-', 'g')
plt.plot(T, P2, '-', 'o')
plt.plot(T, P3, '-', 'r')
plt.plot(T, P4, '-', 'b')
plt.show()
```
