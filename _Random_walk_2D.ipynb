```
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```
def move2D(mi,ni):
    direction = random.choice(['x','y'])
    if direction == 'x':
       r = random.choice([-1,1])
       if r == 1:mf=mi+1
       else:mf=mi-1
       nf=ni
    if direction == 'y':
       r = random.choice([-1,1])
       if r == 1:nf=ni+1
       else:nf=ni-1
       mf=mi
    return mf,nf
```

```
def walk2D(N,pos):
    X=[];Y=[]
    m0=pos[0];n0=pos[1]
    k=0
    while k<N:
        x,y=move2D(m0,n0)
        X.append(x)
        Y.append(y)
        m0=x
        n0=y
        k=k+1
    return X,Y
```

```
N = 1000
U = walk2D(N,[0,0])
V = walk2D(N,[0,0])
W = walk2D(N,[0,0])
P = walk2D(N,[0,0])
Q = walk2D(N,[0,0])
R = walk2D(N,[0,0])
#print('U:',U)
#print('V:',V)
#print(W)
```

```
plt.figure(figsize = [10,10])
plt.plot(U[0],U[1],'o-')
plt.plot(V[0],V[1],'o-')
plt.plot(W[0],W[1],'o-')
plt.plot(P[0],P[1],'o-')
plt.plot(Q[0],Q[1],'o-')
plt.plot(R[0],R[1],'o-')
plt.grid(True)
plt.xlabel('X', fontsize = 20, weight = 'bold')
plt.ylabel('Y', fontsize = 20, weight = 'bold')
plt.title('Brownian motion in 2D')
plt.savefig('Random-2d.png')
```
