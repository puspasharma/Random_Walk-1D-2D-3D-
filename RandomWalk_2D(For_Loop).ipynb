```
import numpy as np
import random as random
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set()
```

```
class walk2D(object):
  def __init__(self, x_initial,y_initial, step, number):
        self.x0 = x_initial
        self.y0 = y_initial
        self.d = step
        self.n = number

  def walk2D(self, xi, yi):   #This function choose a direction and walk one step in 2D
       direction = random.choice(['x','y'])
       if direction == 'x':
           r = random.choice([-1,1])
           xi = xi + r
       if direction == 'y':
           r = random.choice([-1,1])
           yi = yi + r
       return xi,yi
  def walking2D(self):
      X = []
      Y = []
      X.append(self.x0)
      Y.append(self.y0)
      k = 0
      x0 = self.x0
      y0 = self.y0
      while k < self.n:
          x,y = self.walk2D(x0,y0)
          X.append(x)
          Y.append(y)
          x0 = x
          y0 = y
          k = k+1 
      return X,Y
```

```
#Type--1
W = walk2D(0,0,1,1000)
X,Y = W.walking2D()
#len(X)
len(Y)
```

```
plt.figure(figsize = [10,10])
for r in range(500):
    X,Y = W.walking2D()
    plt.plot(X,Y)
plt.xlabel("x-axis", fontsize = 20, color = "b")
plt.ylabel("y-axis", fontsize = 20, color = "b")
plt.grid(True)
plt.title("Walk__2D", fontsize = 25, weight = 'bold', color = 'b')
plt.show()
```
